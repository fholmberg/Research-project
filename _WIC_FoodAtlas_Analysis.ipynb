1) Import the WIC Formula Vendors Excel File

```python
wic_harris_cnty = pd.read_excel('/content/drive/MyDrive/medschool project/WIC_formula_vendors.xlsx')
wic_harris_cnty = wic_harris_cnty[wic_harris_cnty['County'] == 'HARRIS']
wic_harris_cnty = wic_harris_cnty[['County', 'Store Name',
                 'Store Address', 'City', 'Zip',
                 'Tract', 'LILA 1 and 10',
                 'LILA .5 and 10']]
wic_harris_cnty[['Tract', 'Zip']] = wic_harris_cnty[['Tract', 'Zip']].astype(str)
wic_harris_cnty['Tract'] = wic_harris_cnty['Tract'].str[:-2]
wic_harris_cnty = wic_harris_cnty.rename(columns = {'Tract':'CensusTract'})
wic_harris_cnty = wic_harris_cnty[(wic_harris_cnty['Store Name'] != 'ANGEL MEDICAL SUPPLY LLC')]
wic_harris_cnty = wic_harris_cnty[(wic_harris_cnty['Store Name'] != 'FOOD CITY')]
```

2) Import the Food Atlas Dataset

```python
df_food_atlas = pd.read_excel('/content/drive/MyDrive/medschool project/FoodAccessResearchAtlasData2019.xlsx',
                              sheet_name = 'Food Access Research Atlas')
# Filter the data
df_food_atlas = df_food_atlas[(df_food_atlas['County'] == 'Harris County') &(df_food_atlas['State'] == 'Texas')]
df_food_atlas['CensusTract'] = df_food_atlas['CensusTract'].astype(str)
df_food_atlas
```

```python
LILATracts_1And10_is1 = df_food_atlas[['CensusTract',  'LILATracts_1And10']][df_food_atlas['LILATracts_1And10'] == 1]
print(f'There are # number of instances ', len(LILATracts_1And10_is1), 'where LILATracts_1And10 = 1 in the dataset')

LA1and10_is1_L1_is0 = df_food_atlas[['CensusTract',  'LA1and10','LowIncomeTracts']][(df_food_atlas['LA1and10'] == 1) &\
                                                                              (df_food_atlas['LowIncomeTracts'] == 0)]
print(f'There are # number of instances ', len(LA1and10_is1_L1_is0), 'where LA1and10 = 1 and LowIncomeTracts = 0 in the dataset')

LA1and10_is0_L1_is1 = df_food_atlas[['CensusTract',  'LA1and10', 'LowIncomeTracts']][(df_food_atlas['LA1and10'] == 0) &\
                                                                              (df_food_atlas['LowIncomeTracts'] == 1)]
print(f'There are # number of instances ', len(LA1and10_is0_L1_is1), 'where LA1and10 = 0 and LowIncomeTracts = 1 in the dataset')

LA1and10_is0_L1_is0 = df_food_atlas[['CensusTract',  'LA1and10', 'LowIncomeTracts']][(df_food_atlas['LA1and10'] == 0) &\
                                                                              (df_food_atlas['LowIncomeTracts'] == 0)]
print(f'There are # number of instances ', len(LA1and10_is0_L1_is0), 'where LA1and10 = 0 and LowIncomeTracts = 0 in the dataset')
```

# Combine the two datasets
- To add the features 'LA1and10' and 'LAhalfand10' to the WIC Dataset

```python
atlas_wic_combined = pd.merge(wic_harris_cnty[['Store Name', 'Store Address', 'City', 'Zip', 'CensusTract',
       'LILA 1 and 10', 'LILA .5 and 10']],
                              df_food_atlas,
                              on = 'CensusTract',
                              how='inner')
atlas_wic_combined
```
