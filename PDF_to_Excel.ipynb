```python
#import pandas and numpy for data analysis
import pandas as pd
import numpy as np

#import drive to set up Colab and Drive connection to access files in the Drive
from google.colab import drive
import pdfplumber
drive.mount("/content/drive")
```

```python
# Open the PDF file
pdf = pdfplumber.open('/content/drive/MyDrive/medschool project/WIC Vendors.pdf')

# Create an empty DataFrame
df = pd.DataFrame()

# Iterate through the pages of the PDF file
for page in pdf.pages:

    # Extract the table data from the page
    table = page.extract_table()

    # Add the table data to the DataFrame
    df = df.append(table)
```

```python
# Assign the first row of the DataFrame to a variable
first_row = df.iloc[0, :]
# Set the column labels to equal the values in the variable
df.columns = first_row
# Drop the first row from the DataFrame
df = df.iloc[1:, :]
# Remove col titles as rows
df = df[df['County'] != 'County']
df
```

```python
path = '/content/drive/MyDrive/medschool project/WIC_formula_vendors.xlsx' #save the final dataset to drive

# Save the Excel file to your Google Drive folder
df.to_excel(path)
```
